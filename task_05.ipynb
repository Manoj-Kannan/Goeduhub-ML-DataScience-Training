### 4.Data Selection – Indexing and Slicing data
### Indexing:

```python
# indexing using column name
df[['Title','Genre','Description']].head()
```

```python
# indexing using loc()
# loc - Access a group of rows and columns by label(s)
# Syntax: df.loc[row_id, column_id]
df.loc[[0,1,2,3,4],['Title','Genre','Description']]
```

```python
# indexing using iloc
# .iloc[] is primarily integer position based (from 0 to length-1 of the axis)
print('Element in index(0,1):',df.iloc[0,1])
print('Elements in 0th row and all columns:\n',df.iloc[0,:])
```

### Slicing:

```python
# slicing by row-number
df[0:5]
```

```python
# slicing - using loc
df.loc[:, 'Title':'Description'] 
```

```python
# slicing using iloc
df.iloc[0:5, 4:9]
# elements in 0-5 rows and 4-9(index) columns
```

### 5. Data Selection – Based on Conditional filtering

```python
# usual method
print('Movies released after 2012:')
df[df['Year']>2012].head()
```

```python
# using loc
print('Movies released after 2012:')
df.loc[(df['Year']>2012),:].head()
```

```python
# using df.where() - Replace values where the condition is False
# here we just use to locate values
print('Movies released after 2012:')
df.where(df['Year']>2012).head()

# Note: can also use df.query() - to locate values which meets the specified condition
```

### 6.Groupby operations

```python
# df.groupby() - segregates the data by applying a function, and combining the results.
df.groupby('Genre').sum()
```

```python
# groupby() by multiple features
df.groupby(['Genre','Year']).sum()
```

```python
# groupby() by multiple features and analyse a single feature
df.groupby(['Genre','Year'])['Rating'].sum()
```

```python
df.groupby(['Genre','Year'])['Rating'].count()
```

### 8.Dealing with missing values

```python
# check missing values
df.isna().any()
# True - contains missiong values
# False - no missing values
```

```python
# Total number of missing values
print(df.isna().sum())
```

```python
# replace missing values by mean of the feature column
df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean(),inplace=True)
```

```python
# check missing values
df.isna().any()
```
