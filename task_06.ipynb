### 2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.

```python
Days = np.array([1,2,3,4,5,6,7,8])
Speed = np.array([60,62,61,58,56,57,46,63])

#PLot
plt.plot(Days,Speed,'g:') 
plt.scatter(Days,Speed, s=100, marker='+', c = 'r')
plt.title('Car Speed Measurement')
plt.xlabel('days')
plt.ylabel('speed')

# plot - Plot lines and markers to the Axes.
# 'g:'' - string characters are accepted to control the line style or marker
# scatter - A scatter plot of y vs. x with varying marker size and color.
# s - marker size
# marker - marker style
# c - marker color
```

### 3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

days=[1,2,3,4,5,6,7,8]

max_speed=[80,91,92,88,77,79,76,75]

min_speed=[42,43,40,42,33,36,34,35]

avg_speed=[46,58,57,56,40,42,41,36]

```python
days=np.array([1,2,3,4,5,6,7,8])
max_speed=np.array([80,91,92,88,77,79,76,75])
min_speed=np.array([42,43,40,42,33,36,34,35])
avg_speed=np.array([46,58,57,56,40,42,41,36])

#PLot
plt.figure(figsize=(15,5))

#plot 1
plt.subplot(1,3,1)
plt.plot(days,min_speed,'b:')
plt.scatter(days,min_speed, s=150, marker='s', c = 'lightgreen',label= 'min_speed')
plt.title('Days vs Min_Speed')
plt.xlabel('Days')
plt.ylabel('Min_Speed')
plt.legend()
plt.grid()

#plot 2
plt.subplot(1,3,2)
plt.plot(days,max_speed,'c:')
plt.scatter(days,max_speed, s=150, marker='*', c = 'r',label='max_speed')
plt.title('Days vs Max_Speed')
plt.xlabel('Days')
plt.ylabel('Max_Speed')
plt.legend()
plt.grid()

#plot 3
plt.subplot(1,3,3)
plt.plot(days,avg_speed,'y:')
plt.scatter(days,avg_speed, s=150, marker='o', c = 'm',label='avg_speed')
plt.title('Days vs Avg_Speed')
plt.xlabel('Days')
plt.ylabel('Avg_Speed')
plt.legend()
plt.grid()

plt.suptitle('Car Speed Measurement',fontsize=18)

# Adding legend, which helps us recognize the curve according to it's color
# grid() - to add grid lines to the plot.
```

### 4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
X = np.arange(0, math.pi*2, 0.05)

y = np.sin(X)
z = np.cos(X)

plt.plot(X, y, color='r', label='sin')
plt.plot(X, z, color='g', label='cos')

# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine functions")

plt.legend()
```

### 5. Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]
### Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages = np.array(['Python', 'SQL', 'Java', 'C++', 'JavaScript'])
Popularity = np.array([56, 39, 34, 34, 29])
Security = np.array([44 ,36 ,55, 50, 42])
#figure 1
fig1 = plt.figure(figsize=(15,5))
fig1.suptitle('Vertical Bar Chart',fontsize=18)

ax1 = fig1.add_subplot(1,2,1)
ax2 = fig1.add_subplot(1,2,2)

ax1.bar(Languages,Popularity,color= 'lightblue')
ax1.set_title('Languages vs Popularity')
ax1.set_xlabel('Languages')
ax1.set_ylabel('Popularity')
ax2.bar(Languages,Security,color= 'lightgreen')
ax2.set_title('Languages vs Security')
ax2.set_xlabel('Languages')
ax2.set_ylabel('Security')

#figure 2
fig2 = plt.figure(figsize=(15,5))
fig2.suptitle('Horizontal Bar Chart',fontsize=18)

ax1 = fig2.add_subplot(1,2,1)
ax2 = fig2.add_subplot(1,2,2)

ax1.barh(Languages,Popularity,color= 'lightblue')
ax1.set_title('Languages vs Popularity')
ax1.set_xlabel('Languages')
ax1.set_ylabel('Popularity')
ax2.barh(Languages,Security,color= 'lightgreen')
ax2.set_title('Languages vs Security')
ax2.set_xlabel('Languages')
ax2.set_ylabel('Security')
```

### 6.  Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Histogram showing Below Average, Average and Execellent distribution

```python
Marks = np.array([ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])

#PLot
plt.figure(figsize=(8,5))
plt.hist(Marks,bins=[0,20,40,60,80,100],color= 'lightgreen',edgecolor='g')
plt.title('Students Marks',fontsize=18)
plt.xlabel('Marks')
plt.ylabel('No. of Students')
plt.grid()
```

### 7.  Titanic Data Set Download Data

Load the data file

(i) Create a pie chart presenting the male/female proportion 

(ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
df = pd.read_csv('Datasets/titanic.csv')
df.head()
```

```python
df.info()
```

```python
labels = df['Sex'].unique()
values = df['Sex'].value_counts()
print(values)

#PLot
plt.figure(figsize=(8,5))
plt.pie(values, labels = labels, autopct='%1.1f%%', 
        explode=(0.05,0), colors=('lightblue','lightgreen'))
plt.title('Sex Ratio',fontsize=18)
plt.show()
```

```python
male = df[df['Sex'] == 'male']
female = df[df['Sex'] == 'female']
```

```python
#PLot
plt.figure(figsize=(10,6))
plt.scatter(male['Fare'],male['Age'],label='Male')
plt.scatter(female['Fare'],female['Age'],label='Female')
plt.title('Fare vs Age vs Sex',fontsize=18)
plt.xlabel('Fare',fontsize=15)
plt.ylabel('Age',fontsize=15)
plt.legend()
```
